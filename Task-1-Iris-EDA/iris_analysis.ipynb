# Introduction
This notebook explores the Iris dataset to understand its structure, features, and relationships using data visualization techniques.


## Data Loading
The Iris dataset is loaded using the pandas library from an open-source CSV file.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better plot visuals
sns.set(style="whitegrid")
# Load dataset using pandas
iris = pd.read_csv(
    "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
)
```

## Dataset Structure and Preview
This section displays the basic structure of the Iris dataset, including its dimensions, column names, and a preview of the first few rows.

```python
from IPython.display import display

print("Dataset Shape:")
print(iris.shape)

print("\nColumn Names:")
print(list(iris.columns))

print("\nFirst 5 Rows:")
display(iris.head())
```

## Data Visualization

The following visualizations are created using **Matplotlib** and **Seaborn** to analyze relationships, distributions, and variability within the Iris dataset.



### Scatter Plot: Sepal Length vs Petal Length

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="petal_length",
    hue="species"
)
plt.title("Sepal Length vs Petal Length")
plt.show()
```

### Histogram (Distribution of a Feature)


```python
plt.figure(figsize=(8, 5))
plt.hist(iris["sepal_length"], bins=20)
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

### Box Plot (Outliers & Spread)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Petal Length by Species")
plt.show()
```

## Conclusion
The Iris dataset is clean and well-structured with no missing values. Visual analysis shows that petal-related features, especially petal length, are highly effective in distinguishing between different species. This makes the dataset suitable for classification tasks.
